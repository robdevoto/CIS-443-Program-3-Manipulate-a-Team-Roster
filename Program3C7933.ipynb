```python
# Program 3
# CIS 443-01
# Due: 4/6/2022
# Written by: C7933
"""Python program to manipulate the roster of a sports team"""

# function using a sentinal-controlled repetition and if..elif..else statement allowing user to choose task to manipulate roster
def display_menu():
    """Function to display menu and allow user to select options to create and manipulate roster"""
   
    while True:
        print('MENU')
        print('i - Input roster data')
        print('a - Add player')
        print('d - Remove player')
        print('u - Update player rating')
        print('r - Print players above a rating')
        print('p - Print entire roster')
        print('s - Save roster data file')
        print('o - Open roster data file')
        print('q - Quit')
        option = input('Choose an option: ').lower()
        
        if option == 'i':
            create_roster()
            display_menu()
        elif option == 'a':
            add_player(roster)
            display_menu()
        elif option == 'd':
            delete_player(roster)
            display_menu()
        elif option == 'u':
            update_rating(roster)
            display_menu()
        elif option == 'r':
            ratings_above(roster)
            display_menu()
        elif option == 'p':
            print_roster(roster)
            display_menu()
        elif option == 's':
            save_roster(roster)
            display_menu()
        elif option == 'o':
            open_roster(roster)
            display_menu()
        elif option == 'q':
            break
        else:
            print('\nInvalid selection\n')
            display_menu()
        break
    
    
def create_roster():
    """Create a roster with valid jersey numbers and power ratings"""
    # get user input and convert to integer
    roster_input = input("Enter each player's jersey numbers (0-99) and power ratings (1-10) in pairs: ")
    int_roster = [int(item) for item in roster_input.split()]

    # create empty lists
    jersey_numbers = []
    player_rating = []

    # split out keys from input list
    for i in int_roster[::2]:
        jersey_numbers.append(i)

    # split out values from input list
    for i in int_roster[1::2]:
        player_rating.append(i)

    # merge keys and values into dictionary
    prelim_roster = {}
    length = min(len(jersey_numbers),len(player_rating))
    for i in range(length):
        key = jersey_numbers[i]
        value = player_rating[i]
        prelim_roster[key] = value

    # create a list of invalid jersey numbers
    invalid_jerseys = []
    for key in prelim_roster.keys():
        if key < 0 or key > 99:
            invalid_jerseys.append(key)

    # remove invalid jersey numbers and corresponding values     
    for key in invalid_jerseys:
        del prelim_roster[key]

    # create a list of invalid ratings
    invalid_ratings = []
    for value in prelim_roster.values():
        if value < 1 or value > 10:
            invalid_ratings.append(value)

    # remove jersey numbers and corresponding values based on invalid rating creating final roster
    global roster
    roster = {k: v for k, v in prelim_roster.items() if v >= 1 and v <= 10}
    print(f'\nRoster: {roster}\n')
    return roster
        
    
def add_player(roster):
    """Add a player to the roster"""
    new_jersey = int(input("Enter a new player's jersey number to add (0-99): "))
    new_rating = int(input("Enter the new player's power rating (1-10): "))
    
    if 0 <= new_jersey <= 99 and 1 <= new_rating <= 10:
        roster[new_jersey] = new_rating
    print(f'\nUpdated roster: {roster}\n')
    return roster


def delete_player(roster):
    """Delete a player from the roster"""
    del_jersey = int(input("Enter a player's jersey number to delete (0-99): "))
    if del_jersey in roster:
        del roster[del_jersey]
    print(f'\nUpdated roster: {roster}\n')
    return roster


def update_rating(roster):
    """Update the power rating of a player"""
    jersey = int(input("Enter a player's jersey number (0-99): "))
    new_rating = int(input("Enter the player's new power rating (1-10): "))
    if jersey in roster:
        if new_rating >= 1 and new_rating <= 10:
            roster[jersey] = new_rating
    print(f'\nUpdated roster: {roster}\n')
    return roster


def ratings_above(roster):
    """Print the players with a rating above..."""
    rating = int(input("Enter a power rating (1-10): "))
    print(f'\n{"Player":^10}{"Rating":^10}')
    print('--------------------')
    for key, value in sorted(roster.items()):
        if value >= rating:
            print(f'{key:^10}{value:^10}')
    print()

    
def print_roster(roster):
    """Print the roster"""
    print(f'\n{"Player":^10}{"Rating":^10}')
    print('-------------------')
    for key, value in sorted(roster.items()):
        print(f'{key:^10}{value:^10}')
    print()    

    
def save_roster(roster):
    """Save the roster to a csv file"""
    import csv
    with open('teamroster.csv', mode='w', newline='') as teamroster:
        writer = csv.writer(teamroster)
        for player, rating in roster.items():
            writer.writerow([player, rating])
    print(f'\nTeam roster saved\n')        
            
def open_roster(roster):
    """Open the roster"""
    import csv
    try:
        with open('teamroster.csv', 'r', newline='') as teamroster:
            print(f'\n{"Player":^10}{"Rating":^10}')
            print('-------------------')
            reader = csv.reader(teamroster)
            for record in reader:
                player, rating = record
                print(f'{player:^10}{rating:^10}')
            print()
    except FileNotFoundError:
        print('\nThe file name you specified does not exist\n')

        
# call the function to run the program
display_menu()
```
